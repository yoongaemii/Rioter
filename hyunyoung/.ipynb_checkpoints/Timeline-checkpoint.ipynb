# Explore Timeline function
https://developer.riotgames.com/api-methods/#match-v4/GET_getMatchTimeline  
Explore 'getMatchTimeline' function which imports all the events and participants status for each time interval 6000(milisec)

## Structure of MatchFrameDto
Response of get_match_timeline: List[MatchFrameDto]
* timestamp
* participantFrames: Map[String, MatchParticipantFrameDto]	
    * participantId, level, xp
    * teamScore, position
    * currentGold, totalGold
    * minionsKilled, jungleMinionsKilled
* events: List[MatchEventDto]	
    * type: CHAMPION_KILL, WARD_PLACED, WARD_KILL, BUILDING_KILL, ELITE_MONSTER_KILL, ITEM_PURCHASED, ITEM_SOLD, ITEM_DESTROYED, ITEM_UNDO, SKILL_LEVEL_UP, ASCENDED_EVENT, CAPTURE_POINT, PORO_KING_SUMMON
    
### Keys for Each Event Type
(\*게임아이디 하나 넣어서 찾은거라 일반화 안될수도)
* CHAMPION_KILL: type, timestamp, position, killerId, victimId, assistingParticipantIds
* WARD_PLACED: type, timestamp, wardType, creatorId
* WARD_KILL: type, timestamp, wardType, killerId
* BUILDING_KILL: type, timestamp, position, killerId, assistingParticipantIds, teamId, buildingType, laneType, towerType
* ELITE_MONSTER_KILL: type, timestamp, position, killerId, monsterType, monsterSubType
* ITEM_PURCHASED: type, timestamp, participantId, itemId
* ITEM_SOLD: type, timestamp, participantId, itemId
* ITEM_DESTROYED: type, timestamp, participantId, itemId
* ITEM_UNDO: type, timestamp, participantId, afterId, beforeId
* SKILL_LEVEL_UP: type, timestamp, participantId, skillSlot, levelUpType  
ASCENDED_EVENT: No Event Found  
CAPTURE_POINT: No Event Found  
PORO_KING_SUMMON: No Event Found  

### Value with specific form
#### Position
For summoner's rift after 2014/11/12 (Different with Match history minimap and in-game minimap)
* min: { x: 0, y: 0} 
* max: { x: 14820, y: 14881} 

#### TeamId/ParticipantID
team: 100
team: 200
몇 번 participant가 몇 팀에 속하는지 왜 없냐

#### Others  
laneType: 'TOP_LANE', 'MID_LANE', 'BOT_LANE'

```python
from datetime import timedelta

# convert milisecond timestamp into python timedelta object
# for sake of compatiability, convert into only minutes and seconds (without hours)
def convert_timestamp(timestamp):
    minDelta = timestamp // 60000 # 60000 miliseconds == 1 minute
    secDelta = (timestamp % 60000) // 1000 # 1000 miliseconds == 1 second
    return timedelta(minutes = minDelta, seconds = secDelta)

# print(convert_timestamp(12500))
```

## Event Contribution of Participants  
For each participant, find out which event they contribute, whether it be critical or assitant __Excluded__  
All the events related to items(because they seem trivial), skillLevelUp
### Cirtical contribution
* CHAMPION_KILL: type, timestamp, position, __killerId__, victimId, assistingParticipantIds
* WARD_PLACED: type, timestamp, wardType, __creatorId__
* WARD_KILL: type, timestamp, wardType, __killerId__
* BUILDING_KILL: type, timestamp, position, __killerId__, assistingParticipantIds, teamId, buildingType, laneType, towerType
* ELITE_MONSTER_KILL: type, timestamp, position, __killerId__, monsterType, monsterSubType

### Assistant contribution
* BUILDING_KILL: type, timestamp, position, killerId, __assistingParticipantIds__, teamId, buildingType, laneType, towerType


```python
def get_contribution(timeline):
    '''
    input: timeline object
    output: dictionary with each participant id as keys, 
    and the list of events they contributed as values
    '''
    # initialize result dictionary
    # key: participantId, value: list of events they contributed
    contribution = {i: [] for i in range(1, 11)} 

    # type of event: name of key of critical contributor
    critical = {'CHAMPION_KILL': 'killerId', 'WARD_PLACED': 'creatorId', 'WARD_KILL': 'killerId'\
                , 'BUILDING_KILL': 'killerId', 'ELITE_MONSTER_KILL': 'killerId'}
    # type of event: name of key of assistant contributors
    assistant = {'BUILDING_KILL': 'assistingParticipantIds'}

    # flattened list of all events in a match
    events = [item for sublist in [t['events'] for t in timeline] for item in sublist]
    for event in events:
        try: # the event falls into one of critical contribution
            keyName = critical[event['type']]
            contribution[event[keyName]].append(event)
        except: # the events not in critical contribution
            continue    
    
    return contribution
```

```python
contribution = get_contribution(timeline)
```

```python
pprint(contribution)
```
