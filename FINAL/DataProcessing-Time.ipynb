```python
challtimevector = {}


for accid in chall:
    endseason = 1558915200000
    endtime = 1558915200000  #May27 00:00 in unix timestamp milliseconds
    startseason = 1548633600000 #Jan28 00:00 in unix timestamp milliseconds
    halfhour = 1800000   
    #create empty timevector of 7x48 filled with 0s
    dayvec = []
    for j in range(0,48):
        dayvec.append(0)
    timevec = []
    for i in range(0,7):
        timevec.append(dayvec)              
        
    
    history = chall[accid] #list of game details for this account id
    gameindex = 0
    maxgameindex = len(history)
    print(accid)
    print("max:{0}".format(maxgameindex))
    
    #will start from the end, backwards to the start of this season.
    for i in range(0, 17):  #total of 17 weeks to look
        
        for day in range(0,7): #from sunday to monday
            dayindex = 6 - day
            for k in range(0,48):  #48 intervals of 30 minutes
                timeindex= 47- k
                thirtyback = endtime - halfhour
                
                while True:
                    if gameindex>=maxgameindex-1: #break the loop when there are no more games to track
                        break
                        
                    elif history[gameindex]['timestamp'] > endseason: #skip those games that are not in the desired range
                        print("greater than endseason")
                        print(gameindex)
                        gameindex+=1
                            
                    #if a game is present in the current range of time, add weight to that range. 
                    elif history[gameindex]['timestamp'] > thirtyback and history[gameindex]['timestamp'] <= endtime:
                        timevec[dayindex][timeindex] += 1
    
                        gameindex+=1
                    
                    #when none of the conditions is met, skip to the next time interval
                    else:
                        break
                        


                
                endtime -= halfhour
                
    #append the result to the dictionary.
    challtimevector[accid] = timevec
    

#export as pickle
with open('CHALLENGER_new_time_vector.pickle', 'wb') as f:
    pickle.dump(challtimevector, f)
```
